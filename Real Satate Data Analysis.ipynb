### Extracting Age and Price Data from whole data to find  and analys the relationship

## Cleaning our data set for Age vs Price

```python
age_df = age_df.dropna()
age_df[age_df.notnull()]
age_df.tail()
```

## Visualizing Data set 'Age' vs 'Price' in Different Forms'

```python
plt.scatter(age_df['Age'], age_df['Price'], c="c", s=15)
plt.xlabel('Age of the Investor')
plt.ylabel('Price in USD')
plt.title('Age vs Price of the property')
```

### Hypothesis: 1 (based on the above scatter graph)

So here We can determine the level of expense done by different age groups:

    * Most of the investors or buyers are more likely to spent from USD 185000 to USD 260000
    * The Estates ranging from USD 300000 to USD 420000 can be the target area for market expansion for the firm
    * The market above USD 450000 have investments but very few compaired the other two sections

```python
age_list = [i for i in range(18, 100)]
age_groups =list((df['Interval'].dropna()).drop_duplicates())
age_sets={
    25:[],
    35:[],
    45:[],
    55:[],
    65:[],
    90:[]
}
total_sum, price_group =0, []
prct_dist = []
```

```python
age_df = pd.read_excel('age_data.xlsx')
del(age_df['ID'])
age_vs_price = age_df.to_dict('split')
age_vs_price = age_vs_price['data']
```

```python
for i in age_vs_price:
    if i[0] < 26: age_sets[25].append(i[1])
    if i[0] < 36: age_sets[35].append(i[1])
    if i[0] < 46: age_sets[45].append(i[1])
    if i[0] < 56: age_sets[55].append(i[1])
    if i[0] < 66: age_sets[65].append(i[1])
    if i[0] < 90: age_sets[90].append(i[1])
```

```python
for i in age_sets:
    price_group.append(sum(age_sets[i]))
    total_sum = total_sum+(sum(age_sets[i]))
    
for i in range(len(price_group)):
    prct_dist.append((price_group[i]/total_sum)*100)
```

```python
print('Sum of Prices: ',total_sum)
print('\nSum of Prices by age: \n', price_group)
print('\nPercentage Distribution: \n', prct_dist)
```

```python
plt.pie(prct_dist, labels=age_groups, shadow=True, radius=2, startangle=200, explode=(0.5,0.1,0,0,0,0.1), autopct='%1.f%%')
plt.title('Percentage Distriburion\n for \n Real Estate investment\nVS\nDifferenet Age Groups', loc='center',pad=100)
```

## Hypothesis: 2 (based on the above Pie chart)

Based on the above pie chart it is very clear that:
    
    *The least spending age group in Real Estate is 18-25
    *The huge investors are above of 46years in age
    *The potential targets for market expansions can be the age group of people from 30-45
